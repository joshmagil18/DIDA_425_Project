```python
import matplotlib.pyplot as plt
import pandas as pd
menu = pd.read_csv('MenuStat_annual_data_2022.csv')
```

```python
menu.head()
```

```python
menu.columns
```

```python
# Select specific columns
filtered_menu = menu[['food_category', 'calories', 'sodium']]
filtered_menu
```

```python
burgers = menu[menu['food_category'] == 'Burgers']
burgers
```

```python
#delete columns
columns_to_drop = ['serving_size_text', 'serving_size_household', 'calories_text', 
                   'total_fat_text', 'saturated_fat_text', 'trans_fat_text', 
                   'cholesterol_text', 'sodium_text', 'carbohydrates_text', 
                   'dietary_fiber_text', 'sugar_text', 'protein_text', 'notes']

menu_cleaned = menu.drop(columns=columns_to_drop)

# Verify the columns have been removed
menu_cleaned.columns
```

```python
menu=menu_cleaned
menu.head()
```

```python
menu.dtypes
```

```python
# Convert calories column to numeric, forcing errors to NaN
menu['calories'] = pd.to_numeric(menu['calories'], errors='coerce')
```

```python
# Filter for items with more than 500 calories
high_calories = menu[menu['calories'] > 1000]
high_calories
```

```python
high_calories_count = (menu['calories'] > 2000).sum()
high_calories_count
```

```python
# Plotting the distribution of calories for high-calorie items
plt.figure(figsize=(10, 6))

# Create a histogram of calorie values
plt.hist(high_calories['calories'], bins=30, color='salmon', edgecolor='black')

# Add titles and labels
plt.title('Distribution of Calorie Values for High-Calorie Items')
plt.xlabel('Calories')
plt.ylabel('Frequency')

plt.xlim(left=500)
plt.ylim(left=200)
# Display the plot
plt.show()

```

```python
# List of columns to convert to float
columns_to_convert = ['calories', 'total_fat', 'saturated_fat', 'trans_fat', 
                      'cholesterol', 'sodium', 'carbohydrates', 
                      'dietary_fiber', 'sugar', 'protein']

# Convert the columns to float
menu[columns_to_convert] = menu[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Verify the data types
menu.dtypes
```

```python
# Filter for items with more than 500 calories
high_sodium = menu[menu['sodium'] > 3300]
high_sodium
```
